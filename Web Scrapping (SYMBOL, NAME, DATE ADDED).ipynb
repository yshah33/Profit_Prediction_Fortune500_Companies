```python
import requests
from bs4 import BeautifulSoup
import csv

# Define the URL of the Wikipedia page
url = "https://en.wikipedia.org/wiki/List_of_S%26P_500_companies"

# Send an HTTP request to the URL
response = requests.get(url)

# Parse the HTML content
soup = BeautifulSoup(response.text, "html.parser")

# Find the table containing the data
table = soup.find("table", {"class": "wikitable"})

# Create a CSV file and write header row
with open("/content/drive/MyDrive/sp500_companies.csv", "w", newline="") as csv_file:
    writer = csv.writer(csv_file)
    writer.writerow(["Symbol", "Security", "GICS Sector", "GICS Sub-Industry", "Headquarters Location", "Date added", "CIK", "Founded"])

    # Iterate over each row of the table skipping the header row
    for row in table.find_all("tr")[1:]:
        # Extract data from each column of the row
        columns = row.find_all("td")

        # Check if the row has the expected number of columns
        if len(columns) == 9:
            symbol = columns[0].text.strip()
            security = columns[1].text.strip()
            gics_sector = columns[3].text.strip()
            gics_sub_industry = columns[4].text.strip()
            headquarters_location = columns[5].text.strip()
            date_added = columns[6].text.strip()
            cik = columns[7].text.strip()
            founded = columns[8].text.strip()

            # Write data to CSV file
            writer.writerow([symbol, security, gics_sector, gics_sub_industry, headquarters_location, date_added, cik, founded])

print("CSV file created successfully.")
```
