```python
import pandas as pd

# Read CSV files into DataFrames
data_2020 = pd.DataFrame(pd.read_csv('Data_wStocks/rank_name_symbol_2020.csv'))
data_2021 = pd.DataFrame(pd.read_csv('Data_wStocks/rank_name_symbol_2021.csv'))
data_2022 = pd.DataFrame(pd.read_csv('Data_wStocks/rank_name_symbol_2022.csv'))
data_2023 = pd.DataFrame(pd.read_csv('Data_wStocks/rank_name_symbol_2023.csv'))
```

```python
# Remove commas from the 'Profit' column
data_2020['Profit'] = data_2020.loc[:,'Profit'].str.replace(',', '')
data_2021['Profit'] = data_2021.loc[:,'Profit'].str.replace(',', '')
data_2022['Profit'] = data_2022.loc[:,'Profit'].str.replace(',', '')
data_2023['Profit'] = data_2023.loc[:,'Profit'].str.replace(',', '')

# Remove $ from the 'Profit' column
data_2020['Profit'] = data_2020.loc[:,'Profit'].str.replace('$', '')
data_2021['Profit'] = data_2021.loc[:,'Profit'].str.replace('$', '')
data_2022['Profit'] = data_2022.loc[:,'Profit'].str.replace('$', '')
data_2023['Profit'] = data_2023.loc[:,'Profit'].str.replace('$', '')

# Remove () from the 'Profit' column
data_2020['Profit'] = data_2020.loc[:,'Profit'].str.replace(')', '')
data_2021['Profit'] = data_2021.loc[:,'Profit'].str.replace(')', '')
data_2022['Profit'] = data_2022.loc[:,'Profit'].str.replace(')', '')
data_2023['Profit'] = data_2023.loc[:,'Profit'].str.replace(')', '')
data_2020['Profit'] = data_2020.loc[:,'Profit'].str.replace('(', '')
data_2021['Profit'] = data_2021.loc[:,'Profit'].str.replace('(', '')
data_2022['Profit'] = data_2022.loc[:,'Profit'].str.replace('(', '')
data_2023['Profit'] = data_2023.loc[:,'Profit'].str.replace('(', '')

# Convert the 'Profit' column to float
data_2020['Profit'] = data_2020.loc[:,'Profit'].astype(float)
data_2021['Profit'] = data_2021.loc[:,'Profit'].astype(float)
data_2022['Profit'] = data_2022.loc[:,'Profit'].astype(float)
data_2023['Profit'] = data_2023.loc[:,'Profit'].astype(float)

# Remove commas from the 'Revenue' column
data_2020['Revenue'] = data_2020.loc[:,'Revenue'].str.replace(',', '')
data_2021['Revenue'] = data_2021.loc[:,'Revenue'].str.replace(',', '')
data_2022['Revenue'] = data_2022.loc[:,'Revenue'].str.replace(',', '')
data_2023['Revenue'] = data_2023.loc[:,'Revenue'].str.replace(',', '')

# Remove $ from the 'Revenue' column
data_2020['Revenue'] = data_2020.loc[:,'Revenue'].str.replace('$', '')
data_2021['Revenue'] = data_2021.loc[:,'Revenue'].str.replace('$', '')
data_2022['Revenue'] = data_2022.loc[:,'Revenue'].str.replace('$', '')
data_2023['Revenue'] = data_2023.loc[:,'Revenue'].str.replace('$', '')

# Convert the 'Revenue' column to float
data_2020['Revenue'] = data_2020.loc[:,'Revenue'].astype(float)
data_2021['Revenue'] = data_2021.loc[:,'Revenue'].astype(float)
data_2022['Revenue'] = data_2022.loc[:,'Revenue'].astype(float)
data_2023['Revenue'] = data_2023.loc[:,'Revenue'].astype(float)


# Convert the Stocks columns to float
data_2020.loc[:, 'm1-o':'gdp-12'] = data_2020.loc[:, 'm1-o':'gdp-12'].astype(float)
data_2021.loc[:, 'm1-o':'gdp-12'] = data_2021.loc[:, 'm1-o':'gdp-12'].astype(float)
data_2022.loc[:, 'm1-o':'gdp-12'] = data_2022.loc[:, 'm1-o':'gdp-12'].astype(float)
data_2023.loc[:, 'm1-o':'gdp-12'] = data_2023.loc[:, 'm1-o':'gdp-12'].astype(float)


# Convert the columns of string type
data_2020.loc[:, 'Name':'Industry'] = data_2020.loc[:, 'Name':'Industry'].astype(str)
data_2021.loc[:, 'Name':'Industry'] = data_2021.loc[:, 'Name':'Industry'].astype(str)
data_2022.loc[:, 'Name':'Industry'] = data_2022.loc[:, 'Name':'Industry'].astype(str)
data_2023.loc[:, 'Name':'Industry'] = data_2023.loc[:, 'Name':'Industry'].astype(str)
```

```python
# Cleaning the data for the missing value
# Drop rows where any of the specified columns have null values
data_2020_cleaned = data_2020.dropna()
data_2021_cleaned = data_2021.dropna()
data_2022_cleaned = data_2022.dropna()
data_2023_cleaned = data_2023.dropna()

new_un_gdp_data_2020 = list(data_2020_cleaned.loc[1,'ur-1':'gdp-12'])
new_data = list(data_2020_cleaned.loc[:,'m1-o':'m12-c'].mean())
i=1
stock_data_2020 = []
for i in range(24):
    if (i % 2)!=0:
        stock_data_2020.append((new_data[i]+new_data[i-1])/2)

new_2020 = []
new_2020.append(2020)
new_2020 += stock_data_2020 + new_un_gdp_data_2020

new_un_gdp_data_2021 = list(data_2021_cleaned.loc[1,'ur-1':'gdp-12'])
new_data = list(data_2021_cleaned.loc[:,'m1-o':'m12-c'].mean())
i=1
stock_data_2021 = []
for i in range(24):

    if (i % 2)!=0:
        stock_data_2021.append((new_data[i]+new_data[i-1])/2)

new_2021 = []
new_2021.append(2021)
new_2021 += stock_data_2021 + new_un_gdp_data_2021


new_un_gdp_data_2022 = list(data_2022_cleaned.loc[1,'ur-1':'gdp-12'])
new_data = list(data_2022_cleaned.loc[:,'m1-o':'m12-c'].mean())
i=1
stock_data_2022 = []
for i in range(24):

    if (i % 2)!=0:
        stock_data_2022.append((new_data[i]+new_data[i-1])/2)

new_2022 = []
new_2022.append(2022)
new_2022+=stock_data_2022 + new_un_gdp_data_2022

new_un_gdp_data_2023 = list(data_2023_cleaned.loc[1,'ur-1':'gdp-12'])
new_data = list(data_2023_cleaned.loc[:,'m1-o':'m12-c'].mean())
i=1
stock_data_2023 = []
for i in range(24):

    if (i % 2)!=0:
        stock_data_2023.append((new_data[i]+new_data[i-1])/2)

new_2023 = []
new_2023.append(2023)
new_2023+=stock_data_2023 + new_un_gdp_data_2023


# Define columns
columns = ['year', 'stock1', 'stock2', 'stock3', 'stock4', 'stock5', 'stock6', 'stock7', 'stock8', 'stock9', 'stock10', 'stock11', 'stock12', 
           'un-1', 'un-2', 'un-3', 'un-4', 'un-5', 'un-6', 'un-7', 'un-8', 'un-9', 'un-10', 'un-11', 'un-12', 'gdp-1', 
           'gdp-2', 'gdp-3', 'gdp-4', 'gdp-5', 'gdp-6', 'gdp-7', 'gdp-8', 'gdp-9', 'gdp-10', 'gdp-11', 'gdp-12']

all = pd.DataFrame(columns=columns)

all.loc[0] = new_2020
all.loc[1] = new_2021
all.loc[2] = new_2022
all.loc[3] = new_2023

all.head()

```

```python
import matplotlib.pyplot as plt

years = all['year'].unique()

# Plotting the monthly trend for each year
for year in years:
    # Extract stock prices for the current year
    stock_prices_year = all[all['year'] == year].iloc[:, 1:13]
    unemployment_rates_per_year = all[all['year'] == year].iloc[:, 13:25]
    gdp_year = all[all['year'] == year].iloc[:, 25:37]
    
    # Plotting
    fig, ax1 = plt.subplots(figsize=(10, 6))
    ax1.plot(range(1, 13), stock_prices_year.iloc[0], marker='o', linestyle='-', label=f'Stock Prices - {year}')
    ax1.set_xlabel('Month')
    ax1.set_ylabel('Stock Price')

    ax2 = ax1.twinx()
    color = 'tab:red'
    ax2.plot(range(1, 13), gdp_year.iloc[0], marker='o', linestyle='-', label=f'GDP - {year}', color = color)
    ax2.set_xlabel('Month')
    ax2.set_ylabel('GDP')

    plt.title(f'Monthly Trend of Stocks and GDP for Years {year}')
    plt.xticks(range(1, 13))  # Set x-axis ticks to months
    ax1.grid(True)
    # ax1.legend()
    # ax2.grid(True)
    # ax2.legend()
    plt.show()
```
